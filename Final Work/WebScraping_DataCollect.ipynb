```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen("http://www.informatika.bf.uni-lj.si/magus-romance2.html")
bs = BeautifulSoup(html, 'html.parser')
#print(bs.prettify())
```

```python
import sys

tabela = bs.find_all('table')
#print("TABELA 1 ")
#print(tabela[1])
#print("TABELA 1 FIM")
tabela = tabela[1]
linhas = tabela.find_all('tr')
print(type(linhas))
#print(linhas)
#print(sys.getsizeof(linhas))
cont = 0
en = []
en_ready = []
pt = []
pt_ready = []
en_edit = []
pt_edit = []
for i in linhas:
  cont = cont + 1
  filhas = i.findChildren("td")
  #print(sys.getsizeof(filhas))
  #print(cont)
  #print(filhas)

  en.append(filhas[1].text)
  en_edit = [s for s in en if '\n' in s]
  en_ready = [s.replace('\n', '') for s in en_edit]
  en_ready = [s.replace(' ', '') for s in en_ready]
  pt.append(filhas[7].text)
  pt_edit = [s for s in pt if '\n' in s]
  pt_ready = [s.replace('\n', '') for s in pt_edit]
  pt_ready = [s.replace(' ', '') for s in pt_ready]

#print(pt)
print(pt_ready)
#print(en)
print(en_ready)


#resul = map(str.rstrip, pt)
#str_list = filter(None, list(resul))
#print(list(str_list))
```
